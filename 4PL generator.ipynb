```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the 4PL curve
A = 100  # Upper asymptote (maximum response)
D = 10   # Lower asymptote (minimum response)
C = 50   # Inflection point (concentration at the midpoint of the curve)

# Different Hill slopes (steepness of the curve)
B_values = [1, 2, 4]

# Generate ascending X values in the range of 10^3 to 10^6 on a log scale
X = np.logspace(0.1, 4, 500)

# Plot the 4PL curves for each B value
plt.figure(figsize=(8, 6))

for B in B_values:
    # Calculate Y values using the 4PL equation for the current B
    Y = D + (A - D) / (1 + (C / X)**B)
    plt.plot(X, Y, label=f'4PL Curve (B={B})')

# Highlight the inflection point (C) and asymptotes (A and D)
plt.axvline(x=C, color='r', linestyle='--', label=f'Inflection Point (C): {C}')
plt.axhline(y=A, color='g', linestyle='--', label=f'Upper Asymptote (A): {A}')
plt.axhline(y=D, color='b', linestyle='--', label=f'Lower Asymptote (D): {D}')

# Add labels and legend
plt.xscale('log')  # Use log scale for X
plt.xlabel('Concentration (X) - Log Scale')
plt.ylabel('Response (Y)')
plt.title('Four-Parameter Logistic (4PL) Curves with Different Hill Slopes (B)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
